```python
#EDA covert list to str
import ast
df['genres_main'] = df['genres'].apply(lambda x: ast.literal_eval(x) if type(x) == str else x)
```

```python
df['genres_main'] = df['genres_main'].apply(lambda x: ', '.join(x))
```

```python
df.drop(['genres', 'releaseYear', 'numVotes'], axis=1, inplace=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

cv = CountVectorizer(max_features=500, stop_words='english')
```

```python
vector = cv.fit_transform(df['genres_main']).toarray()
```

```python
vector.shape
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
df[df['title'] == 'The Godfather'].index[0]
```

```python
def recommend(movie):
    lis = []
    index = df[df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[:6]:
#         print(df.iloc[i[0]].title)
        if df.iloc[i[0]].title != movie:
            lis.append(df.iloc[i[0]].title)
    return lis
```

### Using both Score and Genre

```python
def recommend_Wscore(movie_name):
    lis = []
    movie = df[df['title'] == movie_name]
    index = movie.index[0]
    score = movie['averageRating'].values[0]
    
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances:
        if df.iloc[i[0]].averageRating >= score:
#             print(df.iloc[i[0]].title)
            if df.iloc[i[0]].title != movie_name:
                lis.append(df.iloc[i[0]].title)
    return lis[:6]
```

```python
recommend('The Godfather')
```

```python
recommend_Wscore('The Godfather')
```
